```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dataset/weather_data_new.csv")
df
```

```python
new_df =df.replace({
    "temperature": -88888,
    "event":"0"
},np.NaN)
new_df
```

```python
new_df = new_df.fillna({
    "event": "no event"
})
new_df
```

```python
new_df = new_df.replace({
    -99999:np.NaN,
    "no event": "Sunny"
})
new_df
```

```python
new_df = new_df.fillna({
    "temperature": "32F",
    "windspeed": "6 mph"
})
new_df
```

```python
new_df = new_df.replace({
    "temperature":"[A-Za-z]",
    "windspeed":"[A-Za-z]"
},"", regex = True)
new_df
```

```python
new_df = new_df.replace(["Sunny", "Rain","Snow"], [1,2,3])
new_df
```
